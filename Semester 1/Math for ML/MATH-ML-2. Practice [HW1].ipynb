````python
# Создадим матрицу наблюдений X и вектор правильных ответов y
X = data.drop(columns=['Well', 'Prod'])  # убираем идентификатор и целевой признак
y = data['Prod']

# Добавляем в X столбец единиц (свободный член)
X_b = np.column_stack([np.ones(X.shape[0]), X])

# Вычисляем коэффициенты методом наименьших квадратов
# (X_b.T * X_b)^(-1) * X_b.T * y
w_hat = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

print('Коэффициенты линейной регрессии (МНК):')
print(np.round(w_hat).astype(int))  # округляем до целого
````
